```python
X = df.iloc[:,[2,3]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (1,10):
    kmeans = KMeans(n_clusters = i, init = "k-means++",random_state=40)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, init="k-means++", random_state=30)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans ==0,0],X[y_kmeans==0,1],s=100,c="blue",label="Cluster 1")
plt.scatter(X[y_kmeans ==1,0],X[y_kmeans==1,1],s=100,c="green",label="Cluster 2")
plt.scatter(X[y_kmeans ==2,0],X[y_kmeans==2,1],s=100,c="red",label="Cluster 3")
plt.scatter(X[y_kmeans ==3,0],X[y_kmeans==3,1],s=100,c="magenta",label="Cluster 4")
plt.scatter(X[y_kmeans ==4,0],X[y_kmeans==4,1],s=100,c="cyan",label="Cluster 5")
plt.scatter(X[y_kmeans ==5,0],X[y_kmeans==5,1],s=100,c="orange",label="Cluster 6")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c="yellow",label="Centroid")
plt.legend()
plt.show()
```
